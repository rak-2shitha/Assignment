#### Bagging vs Boosting

#### Bagging (Bootstrap Aggregating)
 
Trains multiple models independently

Uses random samples (with replacement)

Reduces variance

#### Example: Random Forest

#### Boosting
Models are trained sequentially

Each model corrects previous errors

Focuses on difficult samples

#### Example: AdaBoost
