##### ●	Apply scaling techniques to numerical features:

a.	Standard Scaling  

b. Min-Max Scaling


```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv(r"C:\Users\Rakshitha\Downloads\Data Trasformation\adult_with_headers (1).csv")

# Replace " ?" with NaN
df.replace(" ?", np.nan, inplace=True)

# Separate columns
cat_cols = df.select_dtypes(include='object').columns
num_cols = df.select_dtypes(include=np.number).columns

# Fix: Assign back instead of using inplace

# Categorical → mode
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

# Numerical → median
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

# Check
print(df.isnull().sum())
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Select numerical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns


# a. Standard Scaling

std_scaler = StandardScaler()
df_standard_scaled = df.copy()
df_standard_scaled[num_cols] = std_scaler.fit_transform(df[num_cols])

print("Standard Scaled Data:")
print(df_standard_scaled.head())


# b. Min-Max Scaling

minmax_scaler = MinMaxScaler()
df_minmax_scaled = df.copy()
df_minmax_scaled[num_cols] = minmax_scaler.fit_transform(df[num_cols])

print("\nMin-Max Scaled Data:")
print(df_minmax_scaled.head())
```

##### ●	Discuss the scenarios where each scaling technique is preferred and why.

##### a. Standard Scaling (Z-score Normalization)

What it does:
    
Transforms data so that:

Mean = 0

Standard Deviation = 1

##### Preferred Scenarios

-> When data is normally distributed (Gaussian)

-> When features have different units/scales (e.g., age vs salary)

-> When model assumes centered data

##### Best for Algorithms

- Logistic Regression

- Linear Regression

- Support Vector Machines (SVM)

- Principal Component Analysis (PCA)

##### Why?

- These algorithms depend on distance, variance, or gradients

- Standard scaling ensures all features contribute equally

- Helps in faster convergence during training

##### b. Min-Max Scaling (Normalization)

What it does:
    
Transforms values into a fixed range:

Usually between 0 and 1

##### Preferred Scenarios

-> When data does NOT follow normal distribution

-> When you need bounded values (0 to 1)

-> When working with image or pixel data

##### Best for Algorithms

- K-Nearest Neighbors (KNN)

- Neural Networks (Deep Learning)

- Gradient Descent-based models (sometimes)

##### Why?

-> Keeps data within a fixed range, which helps:
    
- Neural networks perform better

- Distance-based models (like KNN) become more accurate

-> Preserves the original distribution shape

##### ●	Discuss the pros and cons of One-Hot Encoding and Label Encoding.

##### 1. One-Hot Encoding

##### -> Pros

✔ No ordinal relationship

Does not assume any order between categories

✔ Better for nominal data

Works well when categories are independent (e.g., color, city)

✔ Improves model accuracy

Many models perform better without false ranking

##### -> Cons

✖ High dimensionality

Creates many new columns (curse of dimensionality)

✖ Memory and computation cost

Large datasets become heavy

✖ Not suitable for high-cardinality features

Too many categories → too many columns

##### 2. Label Encoding

##### -> Pros

✔ Simple and fast

Easy to implement

✔ Low memory usage

Uses only one column

✔ Good for high-cardinality data

Avoids large number of columns

✔ Works well with tree-based models

Decision Trees, Random Forest handle it well

##### -> Cons

✖ Introduces false order

Model may assume 0 < 1 < 2 (which may not be true)

✖ Not suitable for nominal data

Can reduce model performance

✖ Can mislead distance-based models

Affects algorithms like KNN, SVM

#### 3. Feature Engineering:

##### ●	Create at least 2 new features that could be beneficial for the model. Explain the rationale behind your choices.

```python
# Create copy
df_fe = df.copy()

# Age Group
df_fe['age_group'] = pd.cut(df_fe['age'],
                           bins=[0, 25, 45, 65, 100],
                           labels=['Young', 'Adult', 'Middle_Aged', 'Senior'])

# Work Intensity (FIXED column name)
df_fe['work_intensity'] = pd.cut(df_fe['hours_per_week'],
                                bins=[0, 25, 40, 60, 100],
                                labels=['Part-time', 'Full-time', 'Overtime', 'Workaholic'])

print(df_fe[['age', 'age_group', 'hours_per_week', 'work_intensity']].head())
```

#### Explanation (Rationale)

##### 1. Age Group

Why created?

Raw age is continuous and may not clearly show patterns

Grouping into categories helps identify income trends across life stages

Benefit:

Models can easily learn patterns like:
    
-> Adults/Middle-aged people tend to earn more than younger individuals

##### 2. Work Intensity (hours-per-week)

Why created?

Income is often related to how many hours a person works

Converting into categories makes patterns clearer

Benefit:

Helps model understand:

-> People working overtime or more hours may have higher income
