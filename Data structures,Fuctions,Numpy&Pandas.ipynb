##### 2.	Create a tuple that stores 3 student names. Try changing the second name in the tuple. What happens? Explain why.

```python
students = ("alfiya", "prashanth", "nithin")
print("Original Tuple:", students)
```

```python
students[1]="rakshitha"
students
```

##### Explanation:

A tuple is immutable,

which means: Once created, you cannot change, add, or remove elements

You can convert the tuple to a list, change it, then convert back:

```python
students = ("alfiya", "prashanth", "nithin")

temp = list(students)   # convert to list
temp[1] = "rakshitha"      
students = tuple(temp)  # convert back to tuple

print("Updated Tuple:", students)
```

##### 11.	Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
- Use .info(), .describe()
- Select 'Name' and 'City' columns
- Drop 'City' column
- Fill any missing values in 'Salary' column with the mean
- Remove any duplicate rows


```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Ravi", "Anu", "Kiran", "Megha", "Arun", "Divya", "Ravi", "Sneha", "Rahul", "Anu"],
    "Age": [22, 23, 21, 24, 22, 23, 22, 25, 24, 23],
    "City": ["Bangalore", "Mysore", "Delhi", "Mumbai", "Chennai", "Bangalore", "Bangalore", "Delhi", "Mumbai", "Mysore"],
    "Salary": [25000, 30000, np.nan, 40000, 35000, np.nan, 25000, 45000, 40000, 30000]
}

df = pd.DataFrame(data)

print("Original DataFrame:\n", df)
```

##### -> Use .info(), .describe()

```python
#info 
print("\nInfo:")
print(df.info())

# select columns
print("\nDescribe:")
print(df.describe())
```

##### -> Select 'Name' and 'City' columns
##### -> Drop 'City' column

```python
# select columns
print("\nSelected Columns (Name, City):")
print(df[["Name", "City"]])

# drop city column
df = df.drop("City", axis=1)
print("\nAfter Dropping 'City':\n", df)
```

##### -> Fill any missing values in 'Salary' column with the mean

```python
# fill missing salary with mean
mean_salary = df["Salary"].mean()
df["Salary"].fillna(mean_salary, inplace=True)
print("\nAfter Filling Missing Salary:\n", df)
```

##### -> Remove any duplicate rows

```python
# remove duplicates
df = df.drop_duplicates()
print("\nAfter Removing Duplicates:\n", df)
```
