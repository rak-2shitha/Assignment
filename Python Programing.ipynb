#### Welcome to assignment on Python Programming. In this assignment you are supposed to process/read/work on a file name “Employee.csv”. Which contain employee information of an organization. Following information’s are available for each employee:

•	EID – Employee ID

•	Employee Name

•	Gender

•	Salary

•	Date of Joining

In this assignment you are supposed to write functions for following objective:


```python
import pandas as pd

# Load CSV file
df = pd.read_csv(r"C:\Users\Rakshitha\Downloads\employee.csv")

# 1. Greet employee by first name
def greet_employee(emp_id):
    emp = df[df['EID'] == emp_id]
    if emp.empty:
        return "Employee not found"
    
    name = emp.iloc[0]['Employee Name']
    first_name = name.split()[0]
    gender = emp.iloc[0]['Gender']
    
    title = "Mr" if gender.lower() == "male" else "Ms"
    return f"Hello, {title} {first_name}"


# 2. Average salary of each department
def average_salary_by_department():
    return df.groupby('Department')['Salary'].mean()


# 3. Employee details by ID
def employee_details(emp_id):
    emp = df[df['EID'] == emp_id]
    if emp.empty:
        return "Employee not found"
    
    name = emp.iloc[0]['Employee Name']
    doj = emp.iloc[0]['Date of Joining']
    dept = emp.iloc[0]['Department']
    
    return f"{name} | {doj} | {dept}"


# 4. Sort employees from oldest to newest
def sort_employees_by_doj():
    df['Date of Joining'] = pd.to_datetime(df['Date of Joining'], dayfirst=True)
    return df.sort_values(by='Date of Joining')


# Example calls
print(greet_employee("E123D"))
print(average_salary_by_department())
print(employee_details("E123D"))
print(sort_employees_by_doj())
```

#### •	Function for sorting employees from Oldest to Newest.



```python
import pandas as pd

# Load CSV file
df = pd.read_csv(r"C:\Users\Rakshitha\Downloads\employee.csv")

# Function to sort employees from oldest to newest (by Date of Joining)
def sort_employees_by_doj():
    df['Date of Joining'] = pd.to_datetime(df['Date of Joining'], dayfirst=True)
    sorted_df = df.sort_values(by='Date of Joining')
    return sorted_df

# Example
print(sort_employees_by_doj())
```
