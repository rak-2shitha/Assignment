```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense


# 1. Load Dataset

file_path = r"C:\Users\Rakshitha\Downloads\sonardataset.csv"
df = pd.read_csv(file_path)


# 2. Split Features and Target

X = df.iloc[:, :-1]
y = df.iloc[:, -1]


# 3. Encode Labels

le = LabelEncoder()
y_encoded = le.fit_transform(y)


# 4. Handle Missing Values

X = X.fillna(X.mean())


# 5. Feature Scaling

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


# 6. Train-Test Split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_encoded, test_size=0.2, random_state=42
)


# 7. Build ANN Model

model = Sequential()

# Input + Hidden Layer
model.add(Dense(32, activation='relu', input_shape=(X_train.shape[1],)))

# Additional Hidden Layer
model.add(Dense(16, activation='relu'))

# Output Layer (Binary Classification)
model.add(Dense(1, activation='sigmoid'))


# 8. Compile Model

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)


# 9. Train Model

history = model.fit(
    X_train, y_train,
    epochs=50,
    batch_size=16,
    validation_split=0.1,
    verbose=1
)


# 10. Evaluate Model

loss, accuracy = model.evaluate(X_test, y_test)
print("\nTest Loss:", loss)
print("Test Accuracy:", accuracy)


# 11. Make Predictions

y_pred = (model.predict(X_test) > 0.5).astype("int32")

print("\nSample Predictions:\n", y_pred[:10])
```

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder, StandardScaler
from scikeras.wrappers import KerasClassifier

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam


# 1. Load Dataset

file_path = r"C:\Users\Rakshitha\Downloads\sonardataset.csv"
df = pd.read_csv(file_path)


# 2. Split Features and Target

X = df.iloc[:, :-1]
y = df.iloc[:, -1]


# 3. Encode Target

le = LabelEncoder()
y_encoded = le.fit_transform(y)


# 4. Handle Missing Values

X = X.fillna(X.mean())


# 5. Feature Scaling

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


# 6. Train-Test Split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_encoded, test_size=0.2, random_state=42
)


# 7. Model Builder Function

def create_model(neurons=32, activation='relu', learning_rate=0.001, layers=1):
    model = Sequential()
    
    # Input layer + first hidden layer
    model.add(Dense(neurons, activation=activation, input_shape=(X_train.shape[1],)))
    
    # Additional hidden layers
    for _ in range(layers - 1):
        model.add(Dense(neurons, activation=activation))
    
    # Output layer
    model.add(Dense(1, activation='sigmoid'))
    
    # Compile model
    optimizer = Adam(learning_rate=learning_rate)
    model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    
    return model


# 8. Wrap Model for GridSearch

model = KerasClassifier(model=create_model, verbose=0)


# 9. Define Hyperparameter Grid

param_grid = {
    'model__neurons': [16, 32, 64],
    'model__activation': ['relu', 'tanh'],
    'model__learning_rate': [0.001, 0.01],
    'model__layers': [1, 2],
    'epochs': [20, 50],
    'batch_size': [8, 16]
}

# 10. Grid Search

grid = GridSearchCV(
    estimator=model,
    param_grid=param_grid,
    cv=3,
    verbose=1,
    n_jobs=-1
)

grid_result = grid.fit(X_train, y_train)


# 11. Best Parameters

print("\nBest Parameters Found:\n", grid_result.best_params_)


# 12. Best Model Evaluation

best_model = grid_result.best_estimator_
accuracy = best_model.score(X_test, y_test)

print("\nTest Accuracy of Best Model:", accuracy)
```

##### 4. Evaluation

в—Џ	Employ suitable metrics such as accuracy, precision, recall, and F1-score to evaluate your model's performance.

в—Џ	Discuss the performance differences between the model with default hyperparameters and the tuned model, emphasizing the effects of hyperparameter tuning.


```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix

# Predict on test data
y_pred_probs = best_model.predict(X_test)
y_pred = (y_pred_probs > 0.5).astype(int)

# Calculate metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Display results
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)

# Detailed classification report
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Confusion matrix
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

##### 1.Model with Default Hyperparameters

A model trained with default hyperparameters typically uses:

Predefined number of layers and neurons
Default activation functions (e.g., ReLU)
Default learning rate (e.g., 0.001 in many optimizers)

Performance characteristics:

May achieve moderate accuracy, but not optimal
Can suffer from:
Underfitting (model too simple)
Overfitting (model too complex without tuning)
Precision, recall, and F1-score may be imbalanced
Performance is largely dependent on initial assumptions rather than data-driven optimization

##### 2. Model with Tuned Hyperparameters

The tuned model is obtained using methods like Grid Search or Random Search, which systematically explore combinations of:

Number of hidden layers
Number of neurons
Activation functions
Learning rate
Batch size and epochs

Performance characteristics:

Typically shows improved accuracy
Better generalization on unseen data
More balanced precision and recall
Higher F1-score, indicating better trade-off between false positives and false negatives
